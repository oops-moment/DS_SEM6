```python
import matplotlib.pyplot as plt

# Read data from the file
with open('./performance_data1.txt', 'r') as file:
    lines = file.readlines()

# Extract relevant information for plotting
processes = []
times = []
for line in lines:
    parts = line.split(', ')
    processes.append(int(parts[0].split(': ')[1]))
    times.append(float(parts[2].split(': ')[1].split(' ')[0]))

# Plot the data
plt.plot(processes, times, marker='o', linestyle='-')
plt.xlabel('Number of Processes')
plt.ylabel('Time (seconds)')
plt.title('Performance Scaling')
plt.grid(True)
plt.show()
```
