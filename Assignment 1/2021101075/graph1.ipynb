```python
import matplotlib.pyplot as plt

# Read data from the file
with open('./performance_data1.txt', 'r') as file:
    lines = file.readlines()

# Extract relevant information for plotting
processes = []
times = []
for line in lines:
    parts = line.split(', ')
    processes.append(int(parts[0].split(': ')[1]))
    times.append(float(parts[2].split(': ')[1].split(' ')[0]))

# Plot the data
plt.plot(processes, times, marker='o', linestyle='-')
plt.xlabel('Number of Processes')
plt.ylabel('Time (seconds)')
plt.title('Performance Scaling')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Read data from the file
file_path = './performance_data2.txt'  # Update this with the correct path to your file
with open(file_path, 'r') as file:
    lines = file.readlines()

# Extract processes and time taken from each line
processes = []
time_taken = []

for line in lines:
    process_index = line.find('Processes: ')
    time_index = line.find('Time taken: ')

    if process_index != -1 and time_index != -1:
        processes.append(int(line[process_index + 11:].split(',')[0]))
        time_taken.append(float(line[time_index + 12:].split()[0]))

# Plot the graph
plt.plot(processes, time_taken, marker='o', linestyle='-', color='b')
plt.title('Performance Metric vs. Number of Processes')
plt.xlabel('Number of Processes')
plt.ylabel('Time taken (seconds)')
plt.grid(True)
plt.show()
```
